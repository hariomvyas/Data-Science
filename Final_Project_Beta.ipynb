```python
def plot_df(df, x, y, title="", xlabel='Date', ylabel='Country', dpi=100):
    plt.figure(figsize=(15,4), dpi=dpi)
    plt.plot(x, y, color='tab:red')
    plt.gca().set(title=title, xlabel=xlabel, ylabel=ylabel)
    plt.show()


plot_df(df4, x=df4['Date'], y=df4['US dollar (USD)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
plot_df(df4, x=df4['Date'], y=df4['Euro (EUR)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
plot_df(df4, x=df4['Date'], y=df4['Japanese yen (JPY)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
df4_copy = df4.copy()
```

```python
# df4_copy.columns
cols = ['US dollar (USD)', 'Euro (EUR)', 'Japanese yen (JPY)',
       'Pound sterling (GBP)', 'Canadian dollar (CAD)', 'Swiss franc (CHF)',
       'Indian rupee (INR)', 'Chinese renmimbi (CNY)', 'Turkish lira (TRY)',
       'Saudi riyal (SAR)', 'Indonesian rupiah (IDR)', 'UAE dirham (AED)',
       'Thai baht THB)', 'Vietnamese dong (VND)', 'Egyptian pound (EGP)',
       'Korean won (KRW)', 'Russian ruble (RUB)', 'South African rand (ZAR)',
       'Australian dollar (AUD)']
df4_copy['US dollar (USD)'].mean()
```

```python
for c in cols:
  df4_copy[c] = df4_copy[c] / df4_copy[c].mean()
  # print(c, ": ",df4_copy[c].mean())
  # print(c, ": ",df4_copy[c].min())
```

```python
df4_copy
```

```python
plot_df(df4_copy, x=df4_copy['Date'], y=df4_copy['US dollar (USD)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
plot_df(df4_copy, x=df4_copy['Date'], y=df4_copy['Euro (EUR)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
plot_df(df4_copy, x=df4_copy['Date'], y=df4_copy['Japanese yen (JPY)'], title='Number of US Airline passengers from 1979 to 2022')
```

```python
for c in cols:
  print('Gold Price from 1979 to 2022 in ', c)
  plot_df(df4_copy, x=df4_copy['Date'], y=df4_copy[c], title='Gold Price from 1979 to 2022')
```
